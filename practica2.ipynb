```python
nombres = ["Diego", "Pedro", "Juan" ,"Carmen"]
print(nombres[2])
print(type(nombres))
del nombres[2]
print(nombres)
nombres.insert(3,"Jose")
print(nombres)
print("para ver la posicion donde se encuentra un valor usamos index, Carmen se enecuentra en la posision:" ,nombres.index("Carmen"))
print(nombres)
print("par ver los elementos de la posicion 1 a la 2, pongo un 3 al final!" ,nombres[1:3])
nombres.append("Susana")
print("para ver los elementos de 2 en 2:" ,nombres[::2])
print("para ver los elementos en orden invertido uso -1:" ,nombres[::-1])
# para listar uno a uno los elementos de la lista
for i in nombres:
  print(i)
```
